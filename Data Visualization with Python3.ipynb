```python
from matplotlib import pyplot as plt
```

```python
x = [1,2,3]
y = [1,4,9]
z = [10,5,1]
plt.plot(x, y)
plt.plot(x, z)
plt.title("Test plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["this is y", "this is z"])

plt.show
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
countries = pd.read_csv('countries.csv')
```

```python
countries
```

```python
#comparacion de crecimiento de Population entre China y Estados unidos.
US = countries[countries.country == 'United States']
China = countries[countries.country == 'China']
US
```

```python
#matplotlib
US_visualization = plt.plot(US.year, US.population / 10**6)
China_visualization = plt.plot(China.year, China.population / 10**6) 
plt.title("Comparision")
plt.legend(["United States","China"])
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```

```python
#matplotlib
US_visualization = plt.plot(US.year, US.population / US.population.iloc[0] * 100)
China_visualization = plt.plot(China.year, China.population / China.population.iloc[0] *100) 
plt.title("Comparision")
plt.legend(["United States","China"])
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```
